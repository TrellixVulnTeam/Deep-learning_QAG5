```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-2, 2, (10000, 1))
y = np.random.uniform(-2, 2, (10000, 1))
z = x + y
```

```python
z
```

```python
t = np.arange(-4, 4, 0.01)
plt.plot(t, norm.pdf(t))
#plt.scatter(z, norm.pdf(z, loc=0, scale=1), marker='.', linewidths=0.01)
num_bins = 100
plt.hist(z, num_bins, normed=1, facecolor='red', alpha=0.7)
plt.show()
```

```python
np.var(z)
```

```python
s = np.zeros((10000, 1))
for i in range(100):
    x = np.random.uniform(-2, 2, (10000, 1))
    s += x
print(s)
```

```python
t = np.arange(-40, 40, 0.01)
plt.plot(t, norm.pdf(t))
#plt.scatter(s, norm.pdf(s, loc=0, scale=1), marker='.', linewidths=0.001)
num_bins = 100
plt.hist(s, num_bins, normed=1, facecolor='red', alpha=0.7)
plt.show()
```

```python
np.var(s)
```
